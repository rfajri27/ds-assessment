## Exploratory Data Analysis

### Explore Products Data

#### Identify Top 10 Product Categories 

```python
new_product_df = pd.merge(
    left=products_df, 
    right=product_category_name_translation_df, 
    how="left", 
    left_on="product_category_name", 
    right_on="product_category_name"
)
new_product_df.head(10)
```

```python
sum_product_category_df = new_product_df.groupby("product_category_name_english").product_id.count().sort_values(ascending=False).reset_index()
sum_product_category_df.rename(columns={
    "product_category_name_english": "product_category",
    "product_id": "product_count"
}, inplace=True)
sum_product_category_df.head(15)
```

```python
plt.figure(figsize=(18, 6))
plt.bar(sum_product_category_df["product_category"].head(10), sum_product_category_df["product_count"].head(10))
plt.xticks(fontsize=15, rotation=0)
plt.title("Top 10 Product Categories", loc="center", fontsize=20, pad=20)
plt.tight_layout()
plt.show()
```

#### Identify Top 10 Best Seller Product Categories

```python
new_product_category_df = new_product_df[["product_id", "product_category_name", "product_category_name_english"]]
new_product_category_df.head()
```

```python
new_order_items_df = pd.merge(
    left=new_product_category_df,
    right=order_items_df,
    how="right",
    left_on="product_id",
    right_on="product_id"
)
new_order_items_df.head()
```

```python
sum_order_items_df = new_order_items_df.groupby("product_category_name_english").order_id.count().sort_values(ascending=False).reset_index()
sum_order_items_df.rename(columns={
    "product_category_name_english": "product_category",
    "order_id": "order_count"
}, inplace=True)
sum_order_items_df.head(15)
```

```python
plt.figure(figsize=(22, 6))
plt.bar(sum_order_items_df["product_category"].head(10), sum_order_items_df["order_count"].head(10))
plt.xticks(fontsize=15)
plt.title("Top 10 Best Seller Product Categories", loc="center", fontsize=20, pad=20)
plt.tight_layout()
plt.show()
```

### Explore Orders Data

#### Identify the percentage of delivered and imperfect orders

```python
orders_df["new_order_status"] = orders_df["order_status"].apply(
    lambda x: "delivered order" if x == "delivered" else "imperfect order"
)
orders_df.sample(5)
```

```python
sum_orders_status_df = orders_df.groupby("new_order_status").order_id.count().sort_values(ascending=False).reset_index()
sum_orders_status_df.rename(columns={
    "new_order_status": "order_status",
    "order_id": "order_count"
}, inplace=True)
sum_orders_status_df
```

```python
plt.figure(figsize=(5, 5))
plt.pie(
    x=sum_orders_status_df["order_count"], 
    labels=sum_orders_status_df["order_status"],
    autopct="%.0f%%"
)
plt.title("The percentage of delivered orders", loc="center", fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```

#### Identify the percentage of payment methods that buyer use

```python
sum_order_payments_df = order_payments_df.groupby("payment_type").order_id.count().sort_values(ascending=False).reset_index()
sum_order_payments_df.rename(columns={
    "payment_type": "payment_type",
    "order_id": "order_count"
}, inplace=True)
sum_order_payments_df
```

```python
sum_order_payments_df.loc[5] = ["others", 1529+3]
sum_order_payments_df.drop([3, 4], inplace=True)
sum_order_payments_df
```

```python
plt.figure(figsize=(5, 5))
plt.pie(
    x=sum_order_payments_df["order_count"], 
    labels=sum_order_payments_df["payment_type"],
    autopct="%.0f%%"
)
plt.title("The percentage of each payment method", loc="center", fontsize=15, pad=20)
plt.tight_layout()
plt.show()
```

#### Identify the number of orders per month

```python
monthly_orders_df = orders_df.resample('M', on='order_purchase_timestamp').order_id.count()
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.head()
```

```python
monthly_orders_df.tail()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(
    monthly_orders_df["order_purchase_timestamp"],
    monthly_orders_df["order_id"],
    marker='o', 
    linewidth=2
)
plt.title("Number of orders per month", loc="center", fontsize=20, pad=20)
labels, locations = plt.yticks()
plt.xticks(rotation=45,fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()
```
