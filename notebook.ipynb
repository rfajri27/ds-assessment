## RFM Analysis

RFM analysis is a marketing technique used to quantitatively rank and group customers based on their **recency**, **frequency**, and **monetary** to identify the best customers and perform targeted marketing campaigns.

* **Recency:** How recently has the customer made a transaction?
* **Frequency:** How often do customers order?
* **Monetary:** How much money have customers spent on products on this website/app?

```python
new_orders_df = pd.merge(
    left=new_orders_df, 
    right=order_payments_df, 
    how="left", 
    left_on="order_id", 
    right_on="order_id"
)
new_orders_df.head()
```

```python
rfm_df = new_orders_df.groupby("customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max",
    "order_id": "count",
    "payment_value": "sum"
})
rfm_df.columns = ["customer_unique_id", "max_order_purchase_timestamp", "frequency", "monetary"]
rfm_df.head()
```

```python
rfm_df["max_order_purchase_timestamp"] = rfm_df["max_order_purchase_timestamp"].dt.date
```

```python
recent_date = new_orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_purchase_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_purchase_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.describe()
```

## Customer Segmentation based on RFM Analysis

### Preprocessing the data

```python
rfm_df.hist()
plt.show()
```

As we can see, Frequency and Monetary are highly right-skewed. To fix this, we use fifth root transformation on the data.

```python
rfm_transformed_df = rfm_df.copy()
for column in rfm_transformed_df.columns[1:3]:
    rfm_transformed_df[column] = rfm_transformed_df[column].apply(lambda x: np.power(x, (1/5))) # 5th root
```

```python
rfm_transformed_df.hist(bins=50)
plt.show()
```

#### Scalling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaled_rfm = scaler.fit_transform(rfm_transformed_df.drop('customer_unique_id', axis = 1))

scaled_rfm_df = pd.DataFrame(scaled_rfm, columns = rfm_transformed_df.columns[1:])
scaled_rfm_df
```

### Performing K-Means Clustering

```python
from sklearn.cluster import KMeans

inertia = []

for k in range(1, 10):
    cluster_model = KMeans(n_clusters=k, random_state=24)
    cluster_model.fit(scaled_rfm_df)
    inertia_value = cluster_model.inertia_
    inertia.append(inertia_value)

plt.figure(figsize=(12,6))
plt.plot(range(1, 10), inertia, marker='o', linewidth=2)
plt.title('Inertia Plot', fontsize = 20, pad=10)
plt.xlabel('Number of Clusters', fontsize=15)
plt.ylabel('Inertia', fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.show()
```

Ideal number of clusters for this data is 5

```python
cluster_model = KMeans(n_clusters=5, random_state=24)
cluster_model.fit(scaled_rfm_df)
```

```python
import pickle
import joblib

joblib.dump(cluster_model, "./model/cluster_model.joblib")
```

```python
cluster_rfm_df = rfm_df.copy()
cluster_rfm_df["cluster"] = cluster_model.labels_
cluster_rfm_df.head()
```

### Exploring the clusters
